## string methonds
## case conversion method
- upper(): it converts entire string into upper case
- lower(): it converts entire string into lower case
- title(): it converts the first letter of every word into upper and reminin lower
- swapdown(): it converts upper case into lower case letter and lower to upper
- capitalize(): it converts first character into upper case and remaining into lower case

  

```python
string = "The sky is blue and ocean is blue"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

# substring Index finding
- index(substring): if substring is exists in string,it returns the first occurence of substring position ,otherwise'ValueError'
- find (substring): if substring is exists in string,it returns the first occurence of substring position ,otherwise '-1'


```python
string = "The sky is blue and ocean is blue"
print(string.find('is'))
print(string.find('z'))
print(string.find('is',10))# it start checking from 10th place
print(string.find('e',10))
print(string.find('e',-5))
```

```python
#count(substring): it returns the total non-overlapping occurences of substring
s = "ABCBCBCbbaabb"
print(s.count("BCB"))
print(s.count("B"))
print(s.count("BCB"))
lower_string = s.lower()#it converts to lowercase
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count("z"))
```

```python
# replace(old_stbstring, new_substring): it replace the old sub-string with new sub string 
string = "The sky is blue and ocean is blue"
print(string.replace("blue","Red"))
print(string.replace("e","E",2))#it replaces only first two occurences
print(string.replace("e","E"))
```

### string checking methods
- String checking methods returns the boolen value as result
- isupper(): it returns True, if entire string is in upper case
- islower(): it returns True, if entire string is in lower case
- isalpha(): it returns True, if entire string contains alphabets
- isdigit(): : it returns True, if entire string contains digits
- isalnum(): : it returns True, if entire string contains alphabets or digits
- starts with(substring): : it returns True, if the string start with substing
- ends with(substring) : it returns True, if the string end with substing

```python
print("CODE".isupper())
print("CODE".islower())
print("code".islower())
print("1234".isdigit())
print("123abc".isdigit())
print("code".isalpha())
print("coed123".isalnum())
print("123".isdigit())
print("code".startswith('c'))
```
