```python
# slicing

s = "codegnan"
print(s[2])
print(s[-1])

print(s[2:7])
print(s[-4:])
print(s[1:4:1])
print(s[-4:])
print(s[-7:])
print(s[::-1])
print(s[::-2])
print(s[-3::-1])
```

```python
#string operations
#1. string concatenation (using + operator)
#2. string repetation (using * operator)

s1 = "code"
s2 = "Gnan"
s3 = s1 + s2
print(s3)
print(s1+s2+s3)
print(s1 * 10)
```
